### world bank gdp data

```python
file=r'C:\Users\Dell\Desktop\Machine Learning\Coursera\ds plot\project\country gdp.xls'
# importing excel file
xls=pd.ExcelFile(file)
gdp=xls.parse('Data')
gdp.head()
```

```python
# renaming column names
gdp.rename(columns=gdp.iloc[2,:],inplace=True)
# removing top 2 rows of df
gdp=gdp.iloc[3:,:]
# resetting index values of df
gdp.index=np.arange(0,len(gdp))
# removing first 4 columns
gdp.drop(gdp.columns[:4],inplace=True,axis=1)
# changing the type of remaining columns to int
gdp.columns=gdp.columns.astype(int)
# removing first 30 columns
gdp.drop(gdp.columns[:30],inplace=True,axis=1)
#gdp.head()
```

```python
gdp.iloc[:,:].head()
```

```python
# check for null values corresponding to the BRICS nations
for i in range(len(index)):
    print(gdp.iloc[index[i][0],:].isna().sum())
```

```python
# converting to trillion $
gdp=gdp/10**12
gdp.head()
```

```python
# dropping year 2018 as it contains null values
gdp.columns[-1]
```

### world bank gdp per capita (ppp)

```python
file=r'C:\Users\Dell\Desktop\Machine Learning\Coursera\ds plot\project\ppp per capita.xls'
xls=pd.ExcelFile(file)
ppc=xls.parse('Data')
#gdp.head()
```

```python
# renaming column names
ppc.rename(columns=ppc.iloc[2,:],inplace=True)
# removing top 2 rows of df
ppc=ppc.iloc[3:,:]
# resetting index values of df
ppc.index=np.arange(0,len(ppc))
# removing first 4 columns
ppc.drop(ppc.columns[:4],inplace=True,axis=1)
# changing the type of remaining columns to int
ppc.columns=ppc.columns.astype(int)
# removing first 30 columns
ppc.drop(ppc.columns[:30],inplace=True,axis=1)
#gdp.head()
```

```python
# checking for null values
for i in range(len(index)):
    print(ppc.iloc[index[i][0],:].isna().sum())
```

```python
# since we have only single null values, they can be replaced with row mean  
for i in range(len(index)):
    ppc.iloc[index[i][0],:].fillna(ppc.iloc[index[i][0],:].mean(),inplace=True)
```

```python
ppc.drop(ppc.columns[-1],inplace=True,axis=1)
ppc.columns[-1]
```
